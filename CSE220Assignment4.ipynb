```python
class Patient:
    def __init__(self, id, name, age, bloodgroup, next, prev):
        self.id = id
        self.name = name
        self.age = age
        self.bloodgroup = bloodgroup
        self.next = next
        self.prev = prev


class WRM:
    def __init__(self):
        self.dummy = Patient(None, None, None, None, None, None)
        self.dummy.next = self.dummy
        self.dummy.prev = self.dummy
        self.tail = self.dummy

    def RegisterPatient(self, id, name, age, bloodgroup):
        print(f'\nSuccess\n\n{name} is Registered')
        new = Patient(id, name, age, bloodgroup, None, None)
        temp = self.dummy.prev
        self.dummy.prev = new
        temp.next = new
        new.next = self.dummy
        new.prev = temp

    def ServePatient(self):
        if self.dummy.next != self.dummy:
            print(f'{self.dummy.next.name} is served ')
            self.dummy.next = self.dummy.next.next
        else:
            print('No patient left for serve')

    def CancelAll(self):
        self.dummy.next = self.dummy
        self.dummy.prev = self.dummy

    def CanDoctorGoHome(self):
        if self.dummy.next != self.dummy:
            print('No. Patients left.')
        else:
            print('Yes. The doctor can go home now.')

    def ShowAllPatient(self):
        temp = self.dummy
        temp = 1
        if temp.next == self.dummy:
            print('No patient left')
        else:
            while temp.next != self.dummy:
                print(f'Name of the patient {temp}: {temp.next.name}')
                temp += 1
                temp = temp.next

    def ReverseTheLine(self):
        if self.dummy.next != self.dummy:
            current = self.dummy.prev
            while current != self.dummy:
                print(f'Name of the patient: {current.name}')
                current = current.prev
        else:
            print('No patient left to reverse')


fntn = WRM()
l1 = 0
while l1 < 2:
    print()
    print(f'Welcome')
    print()
    print(f'Choose an Option')
    print()
    print(f'1. Register Patient\n2. Serve Patient\n3. Cancel All\n4. Can Doctor Go Home?\n5. Show details of all patient\n6. Reverse the Line\n7. Exit\n=========================')
    user_input = int(input('Please enter your choice: '))
    if user_input == 1:
        print('Starting the process...\n')
        id = input('\nPlease enter an ID: ')
        name = input('\nPlease enter a name: ')
        age = int(input('\nPlease enter the age: '))
        bloodgroup = input('\nPlease enter the Blood Group: ')
        fntn.RegisterPatient(id, name, age, bloodgroup)
    elif user_input == 2:
        print()
        fntn.ServePatient()
    elif user_input == 3:
        print()
        print('All appointments are canceled')
        fntn.CancelAll()
    elif user_input == 4:
        print()
        fntn.CanDoctorGoHome()
    elif user_input == 5:
        print()
        fntn.ShowAllPatient()
    elif user_input == 6:
        print()
        fntn.ReverseTheLine()
    elif user_input == 7:
        print()
        print('Thank You for using Waiting Room Management\nEXITING...')
        break
    else:
        print()
        print('Please enter a valid option')
```
